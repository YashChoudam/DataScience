```python
# Mathematical operations
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

# union
union_set = set1.union(set2)
print(union_set)

# Intersection
intersection_set = set1.intersection(set2)
print(intersection_set)

set1.intersection_update(set2)
print(set1) 
```

```python
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

# Difference
print(set1.difference(set2))
print(set2.difference(set1))
```

```python
# Symmetric Difference
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

set1.symmetric_difference(set2)
```

```python
# Sets methods
set1 = {1,2,3}
set2 = {4,5,6}

# Is subset
print(set1.issubset(set2))
```

```python
# When we need to identify all the unique elements fo the list

lst = [1,2,3,4,4,5,1,6]
lst2 = set(lst)
print(lst2)
```

```python
# Counting unique words in the text 
text = "In this module we are learning about sets in python"
words = text.split()

# Convert list of words to set to get unique words
unique_words = set(words)
print(unique_words)
print(len(unique_words))
```
