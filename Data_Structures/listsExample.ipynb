Real - World Examples Using Lists in Python


Create A todo List for the tasks to keep track of the tasks 


```python
to_do_list=["Finish Python" , "Finish Pyndatic in python" , "LangChain"]

# Adding Task
to_do_list.append("OpenAI")
to_do_list.append("Ollama")
print(to_do_list)

# Removing finished task
to_do_list.remove("Finish Python")
print(to_do_list)

# Checking if a task exists in the list
if "Ollama" in to_do_list :
    print("Signup on Ollama website")

print("Remaining Tasks")
for task in to_do_list :
    print(f"-{task}")
    
```

```python
# Organizing student cgpa 
cgpa = [8.26, 7.77, 8.9, 8.4, 7.89]

# Adding a new cgpa
cgpa.append(8.01)

# Caluclating the average cgpa
average_cgpa = sum(cgpa) / len(cgpa)
print(f"The average cgpa of the students is : {average_cgpa}")

# Finding the highest and lowest cgpa
highest_cgpa = max(cgpa) 
lowest_cgpa = min(cgpa)

print("Lowest cgpa : ",lowest_cgpa)
print("Highest cgpa : ",highest_cgpa)
```
