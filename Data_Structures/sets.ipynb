```python
# Creating a set
my_set = {1,2,3,4,5}
print(my_set)
print(type(my_set))
```

```python
my_empty_set = set()
print(type(my_empty_set))
```

```python
my_empty_set = set([1,2,3,4,5])
print(my_empty_set)
```

```python
my_set = {1,2,3,4,5}
my_set.add(6)
print(my_set)
```

```python
my_set.remove(3)
print(my_set)
```

```python
# If you try to remove an element from a set , which do not exists in the set , it will throw error
# Therefore use discard
my_set.discard(11)
```

```python
s1 = {1,2,3,4,5}
print(type(s1))
removed_element = s1.pop() # First in first out therefore first set element is returned
print(removed_element)
print(s1)
```

```python
# Clear the set
s2 = {1,2,3,4}
s2.clear()
print(s2)
```

```python
  # Set membership test
s3 = {1,2,3,4,5}
print(3 in s3)
print(7 in s3)
```

```python
# Mathematical operations
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

# union
union_set = set1.union(set2)
print(union_set)

# Intersection
intersection_set = set1.intersection(set2)
print(intersection_set)

set1.intersection_update(set2)
print(set1)
```

```python
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

# Difference
print(set1.difference(set2))
print(set2.difference(set1))
```

```python
# Symmetric Difference
set1 = {1,2,3,4,5,6}
set2 = {4,5,6,7,8,9}

set1.symmetric_difference(set2)
```

```python
# Sets methods
set1 = {1,2,3}
set2 = {4,5,6}

# Is subset
print(set1.issubset(set2))
```

```python
# When we need to identify all the unique elements fo the list

lst = [1,2,3,4,4,5,1,6]
lst2 = set(lst)
print(lst2)
```

```python
# Counting unique words in the text 
text = "In this module we are learning about sets in python"
words = text.split()

# Convert list of words to set to get unique words
unique_words = set(words)
print(unique_words)
print(len(unique_words))
```
