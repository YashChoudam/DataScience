# Basic Syntax : [expressions for item in iterable] 
# With conditiona Logic : [expression for item in iterable and condition] 

```python
#List comprehansions
lst = [x**2 for x in range(1,11,1)]
print(lst)
```

```python
#List of even numbers square between  0 and 10

sqrLst = [x**2 for x in range(0,11,1) if x%2 == 0]
print(sqrLst)
```
