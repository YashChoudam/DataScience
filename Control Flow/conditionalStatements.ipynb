```python
nums = int(input("Enter the number"))
if nums > 0 and nums % 2 == 0 :
    print("The number is positive and even",nums)
elif nums < 0 and nums % 2 == 0 :
    print("The number is negative and even",nums)
elif nums < 0 and nums % 2 != 0 :
    print("The number is negative and odd",nums)
else :
    print("The number is positive and odd",nums)
```
